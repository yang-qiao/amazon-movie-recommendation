# 03B - Learned Baselines
In this notebook, we implement another baseline model for comparison purposes. The algorithm used is the same as the previous baseline model, namely predicted rating = overall average + user bias + item bias. The difference is that here we utilize the BaselineOnly module from the Surprise library, which treats user bias and item bias as parameters to be learned rather than manually computed.

```python
from surprise import Dataset, evaluate, BaselineOnly, Reader

from hyperopt import tpe, fmin, hp, Trials, STATUS_OK

import pickle as pkl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time

sns.set(style='whitegrid')
% matplotlib inline
```

```python
# set file paths and read in samples
data_path = os.path.join('..','..','data')
data_file_100 = os.path.join(data_path, 'reviews_sample_100.csv')
data_file_50 = os.path.join(data_path, 'reviews_sample_50.csv')
data_file_25 = os.path.join(data_path, 'reviews_sample_25.csv')
data_file_10 = os.path.join(data_path, 'reviews_sample_10.csv')

df_100 = pd.read_csv(data_file_100).drop('Unnamed: 0', axis = 1)
df_50 = pd.read_csv(data_file_50).drop('Unnamed: 0', axis = 1)
df_25 = pd.read_csv(data_file_25).drop('Unnamed: 0', axis = 1)
df_10 = pd.read_csv(data_file_10).drop('Unnamed: 0', axis = 1)
```

```python
# use the load_from_df function from Suprise's Dataset module. This function requires the rating scale to be specified.
reader = Reader(rating_scale=(1,5))
sdf_100 = Dataset.load_from_df(df_100[['reviewerID', 'asin', 'overall']], reader=reader)
```

Surprise's BaselineOnly model can be tuned with three hyperparameters:
1. n_epochs: the number of epochs a model trains for prior to termination
2. reg_i: regularization parameter for item biases
3. reg_u: regularization parameter for user biases

To search for optimal hyperparameter values, we eschew grid search for a more efficient and effective Bayesian optimization method using **tree parzen estimators** provided by the Hyperopt library.

The following two sections first define our objective function on which we would like to optimize (3-fold cross-validated MAE) and then feed this objective function into the fmin function from the Hyperopt library to select a set of optimized hyperparameters.

```python
# define the objective function that the fmin module can later optimize on
def test_baseline(params):
    '''Surprise has a BaselineOnly module that finds the best fit for the given prediction algorithm by
       applying regularization, and here we choose ALS as the method used to compute the baseline estimates. '''
    params['method'] = 'als'
    model = BaselineOnly(bsl_options=params)
    sdf_100.split(3, shuffle=True)
    results = evaluate(model, sdf_100, measures = ['mae'], verbose=True)
    mae = np.array(results['mae']).mean()
    return mae
```

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'trials_baseline_opt'), 'rb') as file_in:
        trials = pkl.load(file_in)
    with open(os.path.join(data_path, 'best_baseline_opt.dict'), 'rb') as file_in:
        best = pkl.load(file_in)
else:
    '''set the range of hyperpatameters for the baseline model
       n_epochs: number of iterations the ALS procedure will be performed
       reg_i: regularization parameter for items
       reg_u: regularization parameter for users
    '''
    trials = Trials()
    space = {
        'n_epochs': hp.choice('n_epochs', list(range(5,21))),
        'reg_i': hp.uniform('reg_i', 0, 10),
        'reg_u': hp.uniform('reg_u', 0, 15)
    }
    
    # Choose the Tree-structured Parzen Estimator (TPE) as the algorithm to optimize the objective function
    best = fmin(algo = tpe.suggest,
               fn = test_baseline,
               trials = trials,
               max_evals = 100, # max number of tests
               space = space)
    with open(os.path.join(data_path, 'trials_baseline_opt'), 'wb') as file_out:
        pkl.dump(trials, file_out)
    with open(os.path.join(data_path, 'best_baseline_opt.dict'), 'wb') as file_in:
        pkl.dump(best, file_in)
```

This section below, accompanied by various graphs, visulizes the distribution of the different values each hyperparameter can take and the corresponding loss values.

```python
def val_diagnostic(val_name, trials):
    '''generates scatter plot and histogram of each parameter, as well as a plot on loss values (MAE)'''
    
    vals = np.array([trial['misc']['vals'][val_name] for trial in trials.trials])
    
    '''For hyperparameters that are specified with hp.choice, such as a range of integers,
       Hyperopt returns the index of the tested hyperparameter value rather than the value itself.
       Thus, for hyperparameter ranges that begin at a non-zero point,
       it is necessary to shift the indices to the corresponding values they represent.
    '''
    if val_name == 'n_epochs':
        vals += 5
    elif val_name == 'n_factors':
        vals += 10
        
    ts = [trial['tid'] for trial in trials.trials]
    results = [trial['result']['loss'] for trial in trials.trials]
    
    fig, axes = plt.subplots(1, 3, figsize = (16,4))
    axes[0].scatter(ts, vals)
    axes[0].set(xlabel='iteration', ylabel=val_name)
    axes[1].hist(np.array(vals).squeeze())
    axes[1].set(xlabel=val_name, ylabel='frequency')
    axes[2].scatter(vals, results)
    axes[2].set(xlabel=val_name, ylabel='loss')
    plt.tight_layout()
```

For each hyperparameter, namely number of epochs and regularization parameters for user and item, we make three plots to understand its distribution and performance. The first and second graphs are respectively a scatterplot and a histogram, in order to visualize where the values tend to cluster as the model optimizes. The third graph is a loss vs. values plot that visualizes how MAE changes with different values. The model picks the combination of hyperparameter values where loss is minimized.

In the case of the baseline with ALS, it appears that larger values of **n_epochs** leads to marginally better performance, while no regularization results in the best performance.

```python
for val in trials.trials[0]['misc']['vals'].keys():
    val_diagnostic(val, trials)
```

```python
# optimized hyperparameter values
best_opt = best.copy()
best_opt['n_epochs'] += 10
best_opt['method'] = 'als'
best_opt
```

```python
# evaluate results for 5 iterations using baseline ALS with optimized hyperparameters
sdf_100.split(5)
model_opt = BaselineOnly(bsl_options=best_opt)
results_opt = evaluate(model_opt, sdf_100, measures = ['mae'])
```

After finding the optimized hyperparameters and the best regularization parameters, we find the median MAE of the learned baseline model to be around 0.7168.

```python
# make a box plot of the results
pd.Series(results_opt['mae']).plot(kind = 'box')
```

The optimal hyperparameter values returned by **fmin()** contain regularization values sampled from a uniform distribution, so while the trials show that lower regularization values result in better performance, regularization values of **0** were not actually tested. Thus, we create a new ALS baseline model with the same value for **n_epochs** as that found by **fmin()**, but keep the regularization parameters at **0**. We find that the MAE actually decreases to be around **0.7166**.

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'results_baseline_als.array'), 'rb') as file_in:
        results_no_reg_als = pkl.load(file_in)
else:
    best_no_reg_als = best_opt.copy()
    best_no_reg_als['reg_i'] = 0
    best_no_reg_als['reg_u'] = 0
    with open(os.path.join(data_path, 'best_baseline_als.dict'), 'wb') as file_in:
        pkl.dump(best_no_reg_als, file_in)
    model_no_reg_als = BaselineOnly(bsl_options=best_no_reg_als)
    results_no_reg_als = evaluate(model_no_reg_als, sdf_100, measures = ['mae'])
    with open(os.path.join(data_path, 'results_baseline_als.array'), 'wb') as file_out:
        pkl.dump(results_no_reg_als, file_out)
```

```python
pd.Series(results_no_reg_als['mae']).plot(kind = 'box')
```

We also use SGD instead of ALS to check if choice of optimizer changes the model's performance. With an epoch number of 30, we find that the resulting MAE performs worse with a value at around **0.7211**.

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'results_baseline_sgd.array'), 'rb') as file_in:
        results_no_reg_sgd = pkl.load(file_in)
else:
    best_no_reg_sgd = best_no_reg_als.copy()
    best_no_reg_sgd['method'] = 'sgd'
    best_no_reg_sgd['n_epochs'] = 30
    with open(os.path.join(data_path, 'best_baseline_sgd.dict'), 'wb') as file_in:
        pkl.dump(best_no_reg_sgd, file_in)
    model_no_reg_sgd = BaselineOnly(bsl_options=best_no_reg_sgd)
    results_no_reg_sgd = evaluate(model_no_reg_sgd, sdf_100, measures = ['mae'])
    with open(os.path.join(data_path, 'results_baseline_sgd.array'), 'wb') as file_out:
        pkl.dump(results_no_reg_sgd, file_out)
```

```python
pd.Series(results_no_reg_sgd['mae']).plot(kind = 'box')
```

# Runtime analysis
In this part of the notebook, we run the learned baseline model with different sample sizes in order to explore how runtime scales as sample size increases. For each sample size, we run the baseline model for 1 iteration, and record the difference between start time and end time.

```python
'''run the learned baseline model with als and no regularization parameters on different samplze sizes 
   and calculate runtime.'''
df_list = [df_10, df_25, df_50, df_100]

use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'runtime_baseline_als.list'), 'rb') as file_in:
        runtime_als_list = pkl.load(file_in)
else:
    runtime_als_list = list()
    for df in df_list:
        sdf = Dataset.load_from_df(df[['reviewerID', 'asin', 'overall']], reader=reader)
        
        # use the entire dataset without splitting into folds
        train = sdf.build_full_trainset()
        
        # time training duraction using model with als and no regularization
        start = time.time()
        model_no_reg_als.train(train)
        end = time.time()

        runtime_als_list.append(end-start)

    with open(os.path.join(data_path, 'runtime_baseline_als.list'), 'wb') as file_out:
        pkl.dump(runtime_als_list, file_out)
```

```python
# plot runtime against sample sizes
n_reviews_list = [len(df) for df in df_list]
plt.figure(figsize = (10, 5))
plt.plot(n_reviews_list, runtime_als_list)
plt.scatter(n_reviews_list, runtime_als_list, color = 'red')
plt.ylim(0, 5)
plt.xlim(0, 500000)
plt.xlabel('Number of reviews')
plt.ylabel('Training time (seconds)')
plt.title('Training time vs input size')
plt.tight_layout()
```

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'runtime_baseline_sgd.list'), 'rb')as file_in:
        runtime_sgd_list = pkl.load(file_in)
else:
    runtime_sgd_list = list()
    for df in df_list:
        sdf = Dataset.load_from_df(df[['reviewerID', 'asin', 'overall']], reader=reader)
        train = sdf.build_full_trainset()

        start = time.time()
        model_no_reg_sgd.train(train)
        end = time.time()

        runtime_sgd_list.append(end-start)

    with open(os.path.join(data_path, 'runtime_baseline_sgd.list'), 'wb')as file_out:
        pkl.dump(runtime_sgd_list, file_out)
```

We repeat the process above with the learned baseline model using SGD, and find a similiar result: as sample size gets larger, runtime scales in a relatively linear manner. We also observe that on average, the ALS model runs faster than SGD.

```python
n_reviews_list = [len(df) for df in df_list]
plt.figure(figsize = (10, 5))
plt.plot(n_reviews_list, runtime_sgd_list)
plt.scatter(n_reviews_list, runtime_sgd_list, color = 'red')
plt.ylim(0, 10)
plt.xlim(0, 500000)
plt.xlabel('Number of reviews')
plt.ylabel('Training time (seconds)')
plt.title('Training time vs input size')
plt.tight_layout()
```
