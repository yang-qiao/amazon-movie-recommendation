### baseline_avg
We quickly recalculate the predictions for the baseline avg model here. Notice that becasue this model predicts one rating for all items, we currently don't make top k recommendations using this model. Therefore, this model will be excluded from the comparison plots of precision and reall rates.

```python
def create_cv_folds(in_df, num_folds = 5):
    
    # calculate start index of each cross-validation fold and end index of final fold
    fold_indices = np.linspace(0, in_df.shape[0], num_folds + 1).astype(int)
    
    # randomly shuffle the dataframe
    in_df = in_df.sample(frac=1, random_state=1)
    
    # create list storing each fold
    df_folds = [in_df.iloc[fold_indices[i]:fold_indices[i+1],:] for i in range(num_folds)]
    
    return df_folds
```

```python
if not use_pretrained:
    
    # create 5 cross-validation folds for each sample set
    folds_dict = dict()
    df_names = ['df_10', 'df_25', 'df_50', 'df_100']

    for df, name in zip(df_list, df_names):
        folds_dict[name] = create_cv_folds(df)

    num_folds = 5
    mae_mean_per_sample = list()
    mae_std_per_sample = list()

    # for each sample set, use folds to create 5 cross-validation metrics then aggregate
    for df, name in zip(df_list, df_names):
        mae_list = list()
        precision_list = list()
        recall_list = list()

        # for each fold that serves as the test set, create a corresponding training set
        # then train model and record evaluation metrics
        for i in range(num_folds):
            test_df = folds_dict[name][i]
            train_df = [folds_dict[name][j] for j in range(num_folds) if j != i]
            train_df = pd.concat(train_df, axis = 0)

            # this constitutes 'training' a crude baseline model
            avg_rating = train_df['overall'].mean()

            # calculate the MAE for this fold
            mae = np.abs(test_df['overall'] - avg_rating).mean()
            mae_list.append(mae)

            # calculate the precision and recall for this fold


        # aggregate MAE's from each fold using mean and standard deviation. this is the mean/std MAE for a sample set.
        mae_array = np.array(mae_list)
        mae_mean_per_sample.append(mae_array.mean())
        mae_std_per_sample.append(mae_array.std())

    results_dict['baseline_avg']['mae_mean'] = mae_mean_per_sample
    results_dict['baseline_avg']['mae_std'] = mae_std_per_sample
```

### baseline_manual

Given a set of predicted ratings and actual ratings, the function belows picks top k recommendations for each user in and then calculates the corresponding precision and recall rates. We define precision rate and recall rate to be the following:
- precision rate = items that are both recommended and receive an actual rating of 4 or above / top k recommendation items ranked by predicted rating
- recall rate = items that are both recommended and receive an actual rating of 4 or above / all items that receive an actual rating of 4 or above. 

```python
def calc_precision_recall(preds_df, relevance_threshold = 4.0, k_recommendations = 10):
    '''this function calculates precision and recall rates for a given dataframe of prediction data'''
    user_groups = preds_df.groupby('reviewerID')
    precision_list = list()
    recall_list = list()
    for user, user_df in user_groups:
        # sort the predicted rating in a descending order
        user_df = user_df.sort_values(by = 'est_rating', ascending = False)
        
        # denominator for recall rate is all items that are rated 4 or above by the user
        num_relevant_items = user_df[user_df['actual_rating'] >= 4.0].shape[0]
        
        # take the number of top k recommended items as the denominator for precision rate
        if user_df.shape[0] > k_recommendations:
            user_df = user_df.iloc[0:k_recommendations,:]
        num_recommended_items = user_df.shape[0]
        
        if num_relevant_items == 0:
            precision = 0.0
            recall = 1.0
        else:
            '''numerator of precision and recall are the items (less then k) that are both 
               recommended and receive an actual rating of 4 or higher'''
            num_relevant_recommended_items = user_df[user_df['actual_rating'] >= 4.0].shape[0]
            precision = num_relevant_recommended_items / num_recommended_items
            recall = num_relevant_recommended_items / num_relevant_items
            
        precision_list.append(precision)
        recall_list.append(recall)
    precision_avg = np.mean(precision_list)
    recall_avg = np.mean(recall_list)
    return precision_avg, recall_avg
```

This section below quickly re-calculates predicted ratings using the manual baseline algorithm, namely predicted rating = overall average + user bias + item bias. We then calculate the mean and standard deviation for the mean absolute errors, precision rates, and recall rates.

```python
def bias_baseline(in_df, in_tests = 5):
    
    # accumulate a new rmse on each test
    mae_list = list()
    
    for i in range(in_tests):
        
        shuffled_df = in_df.sample(frac=1, random_state=i)
        split_index = int(shuffled_df.shape[0] * 0.8)
        train_df = shuffled_df.iloc[0:split_index, :]
        test_df = shuffled_df.iloc[split_index:, :]
        
        overall_mean = train_df['overall'].mean()
        
        # calculate item biases (overall_mean - item_mean)
        item_means = train_df.groupby('asin').mean()
        item_biases = item_means['overall'] - overall_mean
        
        # calculate user biases (overall_mean - user_mean)
        user_means = train_df.groupby('reviewerID').mean()
        user_biases = user_means['overall'] - overall_mean
        
        # apply below function to each row in test_df to obtain prediction
        def calc_prediction(row):
            asin = row['asin']
            reviewerID = row['reviewerID']
            
            if reviewerID in user_biases.index:
                user_bias = user_biases[reviewerID]
            else:
                user_bias = 0
                
            if asin in item_biases.index:
                item_bias = item_biases[asin]
            else:
                item_bias = 0
                
            rating_predicted = overall_mean + user_bias + item_bias
            return rating_predicted
        
        # apply below function to each row in test_df to obtain error
        def calc_abs_error(row):
            return abs(row['predicted'] - row['overall'])
        
        #for each element in 'test', calculuate the squared error 
        test_df['predicted'] = test_df.apply(func = calc_prediction, axis = 1)
        errors = test_df.apply(func = calc_abs_error, axis = 1)
        
        # change df column names for compatibility with function calc_precision_recall
        test_df.columns = ['asin', 'reviewerID', 'actual_rating', 'reviewTime', 'est_rating']
        
        # calculate precision and recall
        precision, recall = calc_precision_recall(test_df)
        precision_list.append(precision)
        recall_list.append(recall)
        
        # calculate rmse and append
        mae = errors.mean()
        mae_list.append(mae)
        
    return np.array(mae_list), np.array(precision_list), np.array(recall_list)

if not use_pretrained:

    mae_mean_per_sample = list()
    mae_std_per_sample = list()
    precision_mean_per_sample = list()
    precision_std_per_sample = list()
    recall_mean_per_sample = list()
    recall_std_per_sample = list()

    # for each sample set, use folds to create 5 cross-validation metrics then aggregate
    for df, name in zip(df_list, df_names):

        mae_array, precision_array, recall_array = bias_baseline(df)

        # aggregate MAE's from each fold using mean and standard deviation. this is the mean/std MAE for a sample set.
        mae_mean_per_sample.append(mae_array.mean())
        mae_std_per_sample.append(mae_array.std())

        # aggregate precision and recall from each fold for precision/recall per sample set
        precision_mean_per_sample.append(precision_array.mean())
        precision_std_per_sample.append(precision_array.std())
        recall_mean_per_sample.append(recall_array.mean())
        recall_std_per_sample.append(recall_array.std())

    results_dict['baseline_man']['mae_mean'] = mae_mean_per_sample
    results_dict['baseline_man']['mae_std'] = mae_std_per_sample
    results_dict['baseline_man']['precision_mean'] = precision_mean_per_sample
    results_dict['baseline_man']['precision_std'] = precision_std_per_sample
    results_dict['baseline_man']['recall_mean'] = recall_mean_per_sample
    results_dict['baseline_man']['recall_std'] = recall_std_per_sample
```

This section below creates a general function that generates evaluation results for all models. Given a model and a list of sample sets that differ in size, the function performs the following tasks:
1. for each sample set, split input dataset into 5 folds for cross-validation;
2. for each test-train pair, calculate MAE, precision rates, and recall rates;
3. calculate the mean and standard deviation of the MAE's, precision rates, and recall rates, and store these metrics in a results dictionary.

```python
def generate_results(model, model_name, df_list, results_dict, num_folds = 5):
    mae_mean_per_sample = list()
    mae_std_per_sample = list()
    precision_mean_per_sample = list()
    precision_std_per_sample = list()
    recall_mean_per_sample = list()
    recall_std_per_sample = list()
    coverage_list = list()

    # for each sample set, use folds to create 5 cross-validation metrics then aggregate
    for df, name in zip(df_list, df_names):
        mae_list = list()
        precision_list = list()
        recall_list = list()

        # for each fold that serves as the test set, create a corresponding training set
        # then train model and record evaluation metrics
        sdf = Dataset.load_from_df(df[['reviewerID', 'asin', 'overall']], reader=reader)
        sdf.split(num_folds)
        for train_sdf, test_sdf in sdf.folds():

            # train model
            model.train(train_sdf)

            # calculate the MAE for this fold
            predictions = model.test(test_sdf)
            abs_error_list = np.abs([predictions[i].est - predictions[i].r_ui for i in range(len(predictions))])
            mae = abs_error_list.mean()
            mae_list.append(mae)

            # calculate precision and recall for this fold
            preds_df = predictions_to_df(predictions)
            precision_mean, recall_mean = calc_precision_recall(preds_df=preds_df)
            precision_list.append(precision_mean)
            recall_list.append(recall_mean)
            
        # aggregate MAE's from each fold using mean and standard deviation. this is the mean/std MAE for a sample set.
        mae_array = np.array(mae_list)
        mae_mean_per_sample.append(mae_array.mean())
        mae_std_per_sample.append(mae_array.std())

        # aggregate precision and recall from each fold using mean and standard deviation.
        # this is the mean/std precision and recall for a sample set.
        precision_mean_per_sample.append(np.mean(precision_list))
        precision_std_per_sample.append(np.std(precision_list))
        recall_mean_per_sample.append(np.mean(recall_list))
        recall_std_per_sample.append(np.std(recall_list))

    results_dict[model_name]['mae_mean'] = mae_mean_per_sample
    results_dict[model_name]['mae_std'] = mae_std_per_sample
    results_dict[model_name]['precision_mean'] = precision_mean_per_sample
    results_dict[model_name]['precision_std'] = precision_std_per_sample
    results_dict[model_name]['recall_mean'] = recall_mean_per_sample
    results_dict[model_name]['recall_std'] = recall_std_per_sample
```

We then apply the generate_result function on the models below with four different sample sets processed for each of the models: baseline sgd, baseline als, svd, nmf, and knn.

```python
if not use_pretrained:

    # baseline_sgd
    model = BaselineOnly(bsl_options=params_dict['baseline_sgd'])
    generate_results(model=model,
                     model_name='baseline_sgd',
                     df_list=df_list,
                     results_dict=results_dict,
                     num_folds = 5)

    # baseline_als
    model = BaselineOnly(bsl_options=params_dict['baseline_als'])
    generate_results(model=model,
                     model_name='baseline_als',
                     df_list=df_list,
                     results_dict=results_dict,
                     num_folds = 5)

    # svd
    model = SVD(**params_dict['svd'])
    generate_results(model=model,
                     model_name='svd',
                     df_list=df_list,
                     results_dict=results_dict,
                     num_folds = 5)

    # nmf
    model = NMF(**params_dict['nmf'])
    generate_results(model=model,
                     model_name='nmf',
                     df_list=df_list,
                     results_dict=results_dict,
                     num_folds = 5)

    # knn
    model = KNNWithZScore(k = params_dict['knn']['k'],
                                  min_k = params_dict['knn']['min_k'],
                                  sim_options = dict(name=params_dict['knn']['name'],
                                                     min_support=5,
                                                     user_based=False))
    generate_results(model=model,
                     model_name='knn',
                     df_list=df_list,
                     results_dict=results_dict,
                     num_folds = 5)
```

# Results
Here we show multiplie plots to compare each model's performance metrics. We show each model's metric on a normal scale on the left and an amplified version on the right.

```python
def plot_metric(metric_name, results_dict, colors, std_fill=1):
    
    metric_mean = metric_name + '_mean'
    metric_std = metric_name + '_std'
    
    ylabel_dict = dict(mae='Mean absolute error', precision='Precision', recall='Recall')
    
    fig, axes = plt.subplots(1,2, figsize = (16,5))
    for model_name, color in zip(results_dict.keys(), colors):
        if model_name != 'baseline_avg' or metric_name not in ['precision', 'recall']:
            means = np.array(results_dict[model_name][metric_mean])
            stds = std_fill * np.array(results_dict[model_name][metric_std])
            axes[0].plot(input_sizes, means, color=color, label=model_name)
            axes[0].fill_between(input_sizes, means+stds, means-stds, color=color, alpha=0.25)

            axes[1].plot(input_sizes, means, color=color, label=model_name)
            axes[1].fill_between(input_sizes, means+stds, means-stds, color=color, alpha=0.25)
        
    axes[0].set(xticks = [0] + input_sizes, xticklabels = [0] + input_sizes, xlim = (0, input_sizes[-1]))
    axes[0].set(ylim = (0, 1), xlabel = 'Number of ratings', ylabel = ylabel_dict[metric_name])
    axes[0].legend()

    axes[1].set(xticks = [0] + input_sizes, xticklabels = [0] + input_sizes, xlim = (0, input_sizes[-1]))
    axes[1].set(xlabel = 'Number of ratings', ylabel = ylabel_dict[metric_name])
    axes[1].legend()
    
    xticklabels0 = axes[0].get_xticklabels()
    xticklabels1 = axes[1].get_xticklabels()
    for xticklabel0, xticklabel1 in zip(xticklabels0, xticklabels1):
        xticklabel0.set(rotation=90)
        xticklabel1.set(rotation=90)

    plt.tight_layout()
```

# Coverage
In the last part of this evaluation notebook we analyze how coverage changes between different models.

```python
def prediction_row_to_tuple(pred_row):
    return (pred_row.uid, pred_row.iid, pred_row.r_ui, pred_row.est)

def predictions_to_df(preds):
    df = pd.DataFrame([prediction_row_to_tuple(preds[i]) for i in range(len(preds))])
    df.columns = ['asin', 'reviewerID', 'actual_rating', 'est_rating']
    return df

def get_top_k(model, df, user, k=10):
    predictions_list = list()
    item_list = df['asin'].unique()
    exclude_list = df[df['reviewerID'] != user]['asin']
    for item in item_list:
        if item not in exclude_list:
            # predict a rating for items all items not already rated by given user
            predictions_list.append(model.predict(user, item, verbose=False))
    predictions_list = sorted(predictions_list, reverse=True, key=lambda x: x.est)
    
    # select the top k items with highest ratings
    predictions_list = predictions_list[:k]
    print(predictions_list)
    recommendations_list = [i.iid for i in predictions_list]
    return recommendations_list

def calc_coverage(model, df, k=10):
    recommended_set = set()
    user_list = df['reviewerID'].unique()
    sdf = Dataset.load_from_df(df[['reviewerID','asin','overall']], reader=reader)
    sdf_train = sdf.build_full_trainset()
    model.train(sdf_train)
    for user in user_list:
        # recommend k movies for each user in the input dataset and find the union of all recommended movies
        new_recs = get_top_k(model=model, df=df, user=user, k=k)
        recommended_set = recommended_set.union(set(new_recs))
    # coverage rate = all recommended movies / all movies available in the input dataset
    coverage = len(recommended_set) / df['asin'].nunique()
    return coverage
```

```python
# feed in the optimal parameters for each model
model_baseline_sgd = BaselineOnly(bsl_options=params_dict['baseline_sgd'])

model_baseline_als = BaselineOnly(bsl_options=params_dict['baseline_als'])

model_svd = SVD(**params_dict['svd'])

model_nmf = NMF(**params_dict['nmf'])

model_knn = KNNWithZScore(k = params_dict['knn']['k'],
                              min_k = params_dict['knn']['min_k'],
                              sim_options = dict(name=params_dict['knn']['name'],
                                                 min_support=5,
                                                 user_based=False))
model_dict = dict(baseline_sgd = model_baseline_sgd,
                  baseline_als = model_baseline_als,
                  svd = model_svd,
                  nmf = model_nmf,
                  knn = model_knn)
```
